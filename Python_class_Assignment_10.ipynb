**Python program to interchange first and last elements in a list.**

```python
def swaplist(newlist):
  newlist[0],newlist[-1] = newlist[-1],newlist[0]
  return newlist


newlist= [10,62,92,83,92,83,76]
swaplist(newlist)
```

**Python program to swap two elements in a list**



```python
def swapList(sl,pos1,pos2):
    n = len(sl)     
    # Swapping 
    temp = sl[pos1]
    sl[pos1] = sl[pos2]
    sl[pos2] = temp
    return sl
```

**Python program to count occurrences of an element in a list**

```python

def count(lst, x):
    count = 0
    for ele in lst:
        if (ele == x):
            count = count + 1
    return count
 
 
lst = [18,18,67,76,67,67,67]
x = 67
print(x,count(lst, x))
```

**What are the ways to find the length of a list**

```python
#Using the len() method to get the length of a list
len(list1)

#Using the length_hint() method to get the length of a list
from operator import length_hint 

size = length_hint(list1)
print(size)

#Using a for loop to get the length of a list
counter = 0
for item in list1:
  counter += 1

print(counter)
```

**Python program to find the smallest and largest number in a list**

```python

def min_max(list1):
    length = len(list1)
    list1.sort()
    print("Largest element is:", list1[length-1])
    print("Smallest element is:", list1[0])

min_max(list1)
```
